```python
import torch
from transformers import BitsAndBytesConfig
from langchain import HuggingFacePipeline
from langchain import PromptTemplate, LLMChain
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
)
```

```python
model_4bit = AutoModelForCausalLM.from_pretrained( "mistralai/Mistral-7B-Instruct-v0.1", device_map="auto",quantization_config=quantization_config, )
tokenizer = AutoTokenizer.from_pretrained("mistralai/Mistral-7B-Instruct-v0.1")
```

```python
import ast
# Convert the string to a list
list1 = ast.literal_eval(list1_str)
list2 = ast.literal_eval(list2_str)

# Remove 'nan' entries from the lists
list1 = [intent for intent in list1 if intent != 'nan']
list2 = [intent for intent in list2 if intent != 'nan']
```

```python
!pip install --upgrade --quiet  rank_bm25

import re
from langchain_community.retrievers import BM25Retriever

# Tokenize intents for BM25
tokenized_intents1 = [re.sub(r'-', ' ', intent).lower() for intent in list1]
tokenized_intents2 = [re.sub(r'_', ' ', intent).lower() for intent in list2]

# Initialize BM25
bm25_1 = BM25Retriever.from_texts(tokenized_intents1)
bm25_2 = BM25Retriever.from_texts(tokenized_intents2)
```

```python
from langchain.prompts import PromptTemplate

template = """[INST] You are a respectful and helpful assistant. Given the dialogue "{utterance}",
select the most appropriate intent from the following options: {most_similar_intent}. Provide only
the intents that best matches the dialogue and do not include any additional text or explanation.[/INST]
"""

# Define the function to generate intents using BM25Retriever
def generate_intent(utterance, bm_type, model, tokenizer):
    # Tokenize the user utterance (removing punctuation, lowercasing, splitting)
    dialogue = re.sub(r'[^\w\s]', '', utterance).lower()
    tokenized_dialogue = dialogue.split()

    # Calculate BM25 scores for the dialogue against all intents
    scores = bm_type.get_relevant_documents(" ".join(tokenized_dialogue))

    if scores:  # Ensure there are results
        # Get the most relevant intent from the BM25 results
        most_similar_intent_list = [score.page_content for score in scores]  # Get list of relevant intents
        most_similar_intent = ", ".join(most_similar_intent_list)  # Convert list to comma-separated string

        # Use the PromptTemplate to generate a model prompt
        prompt = PromptTemplate(template=template, input_variables=["utterance", "most_similar_intent"])

        # Format the prompt with the utterance and most similar intents
        formatted_prompt = prompt.format(
            utterance=utterance,
            most_similar_intent=most_similar_intent
        )

        # Tokenize the prompt and generate a response
        inputs = tokenizer(formatted_prompt, return_tensors="pt").to(model.device)
        outputs = model.generate(**inputs, max_new_tokens=50, temperature=0.1)
        full_response = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

        # Strip off unnecessary parts and return only the intent
        refined_intent = full_response.split('[/INST]')[-1].strip()

        # Strip off any unnecessary parts like asterisks or formatting
        refined_intent = refined_intent.replace('*', '').strip()  # Remove any asterisks

        refined_intent = ", ".join([intent.strip() for intent in refined_intent.splitlines() if intent])

        return refined_intent
    else:
        return "Unknown Intent"

# Example usage
sample_dialogue = "help me find train"
generated_intent = generate_intent(sample_dialogue, bm25_1, model_4bit, tokenizer)
print(f"Generated Intent for '{sample_dialogue}': {generated_intent}")
```

```python
import pandas as pd

# Load the Excel file
file_path = "update_with_intents_RAG_sparse.xlsx"
df = pd.read_excel(file_path)

# Filter rows where turn_id is even (user dialogues)
user_dialogues = df[df['turn_id'] % 2 == 0]

start_row = 0
# Calculate the corresponding index in user_dialogues
start_index = user_dialogues.index[user_dialogues.index >= start_row][0]

# Iterate over the user dialogue rows and fill in intent1 and intent2
for index, row in user_dialogues.loc[start_index:].iterrows():
    utterance = row['utterance']

    # Generate intent
    generated_intent = generate_intent(utterance, bm25_1, model_4bit, tokenizer)

    # Update the DataFrame with the generated intent and doc_id
    df.at[index, 'intent1'] = generated_intent

    # Optionally, save progress periodically
    if (index - start_index) % 10 == 0:
        df.to_excel(file_path, index=False)
        print(f"Processed up to row {index}")

# Final save
df.to_excel(file_path, index=False)
print(f"Excel file '{file_path}' updated with generated intents.")
```
